### **Bookings are more on weekdays or weekends?**

```python

# Calculating the total stays for weekend nights and week nights
total_weekend_nights = df1['stays_in_weekend_nights'].sum()
total_week_nights = df1['stays_in_week_nights'].sum()


# Comparing the totals

if total_week_nights > total_weekend_nights:
  print(f'Week days stays are higher: {total_week_nights}')
else:
  print(f'Weekend stays are higher: {total_weekend_nights}')

# Data for plotting
x = ['Weekend Nights', 'Week Nights']
y = [total_weekend_nights, total_week_nights]

# Creating the bar chart
plt.figure(figsize=(8, 6))
bar=plt.bar(x, y, color=['blue', 'green'])

for bar in bar:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 5000, int(yval), ha='center', va='bottom')

plt.xlabel('Type of Stay')
plt.ylabel('Total Stays')
plt.title('Comparison of Total Stays on Weekends vs. Weekdays')
plt.show()
```

CONCLUSION: The bookings are more on week days thaan weekends.

### **ADR by month**

```python

# Converting the 'arrival_date_month' column to a categorical type with proper order
months_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
df1['arrival_date_month'] = pd.Categorical(df1['arrival_date_month'], categories=months_order, ordered=True)


# Group by 'arrival_date_month' and calculating the average ADR
monthly_adr = df1.groupby('arrival_date_month')['adr'].mean().reset_index()


plt.figure(figsize=(10, 6))
plt.plot(monthly_adr['arrival_date_month'], monthly_adr['adr'], marker='o', linestyle='-')
plt.xticks(monthly_adr['arrival_date_month'], months_order, rotation=45)
plt.xlabel('Month')
plt.ylabel('Average Daily Rate (ADR)')
plt.title('Average Daily Rate (ADR) by Month')
plt.grid(True)
plt.tight_layout()
plt.show()
```

CONCLUSION:

â–¶ **Seasonal trends**:

â–ª The "**August**" month has the highest ADR. This is peak tourist month with special events that drive up demands.

â–ªThe "**January**" month has lowest ADR. This is off-peak tourist month with lower demand.

â–¶**Price fluctuations**: The ADR gradually increases from January to August and then sharply declines from August to November, with a slight uptick in December.








**Count & proportion of bookings**

```python


hotel_counts=df1.groupby( 'hotel').size().reset_index(name='num_bookings')

total_bookings = hotel_counts['num_bookings'].sum()
hotel_counts['Proportion'] = hotel_counts['num_bookings'] / total_bookings

print(hotel_counts)

plt.figure(figsize=(8, 8))

plt.pie(hotel_counts['Proportion'], labels=hotel_counts['hotel'], autopct='%1.1f%%',explode=(0.00,0.09), colors=['skyblue', 'lightgreen'])
plt.title('Proportion of Bookings Over Hotel Type')
plt.show()
```

CONCLUSION: City Hotel is more popular among the customers as they've highest number of bookings.

**Distribution of cancellations over week numbers**

```python
df1['length_of_stay'] = df1['stays_in_weekend_nights'] + df1['stays_in_week_nights']
#print(df[['stays_in_weekend_nights', 'stays_in_week_nights', 'length_of_stay']].head())

#Calculating the average ADR for each length of stay
adr_by_length_of_stay = df1.groupby('length_of_stay')['adr'].mean().reset_index(name='avg adr')

print(adr_by_length_of_stay)




non_zero_df = adr_by_length_of_stay[adr_by_length_of_stay['avg adr'] > 0]

optimal_length_of_stay = adr_by_length_of_stay.loc[non_zero_df['avg adr'].idxmin()]
print(f"The optimal length of stay is {optimal_length_of_stay['length_of_stay']} nights.")


plt.figure(figsize=(12, 6))
sns.lineplot(data=adr_by_length_of_stay, x='length_of_stay', y='avg adr', marker='o', color='blue')
plt.axvline(optimal_length_of_stay['length_of_stay'], color='red', linestyle='--', label=f'Optimal Length of Stay: {optimal_length_of_stay["length_of_stay"]} nights')
plt.xlabel('Length of Stay (nights)')
plt.ylabel('Average Daily Rate (ADR)')
plt.title('Average Daily Rate (ADR) by Length of Stay')
plt.legend()
plt.show()

```

### **Does the guest get the rooms which they reserved?**

```python
df['room_type_match']=df['reserved_room_type']==df['assigned_room_type']

# Counting the number of matches and mismatches
match_counts = df['room_type_match'].value_counts()
print(match_counts)

# Group by hotel type and calculate the match rate
match_rate_by_hotel = df.groupby('hotel')['room_type_match'].mean().reset_index()
match_rate_by_hotel.columns = ['Hotel', 'Match_Rate']

# Printing the match rate for each hotel type
print(match_rate_by_hotel)
```

CONCLUSION: The "**City hotel**" is better at assigning guests the room type they reserved.
